```python
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
```

```python
dir_path = "../data/"
file_path = listdir(dir_path)[0]
data = pd.read_csv(dir_path+file_path, skiprows=12)
data.head()
```

```python
data = data.drop('Unnamed: 11', axis=1)
```

```python
len = data.shape[0]
for col in data.columns:
    print(col, '{}%'.format(data[col].isna().sum()/len*100))
```

```python
pd.to_datetime([data['Total'], format='%H:%M:%S'])
```

```python
t = pd.to_timedelta(data['Total']).astype('timedelta64[s]')
t[t < 900000].plot(kind='hist')
```
